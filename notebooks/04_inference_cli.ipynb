# Inference CLI Examples

This notebook demonstrates how to use the trained models for inference via the CLI interface.

```python
import pandas as pd
from pathlib import Path
import sys
import subprocess

# Set project root
project_root = Path().resolve().parent
sys.path.insert(0, str(project_root))
```

## Domain Prediction Examples

```python
# Example: Predict a domain using CLI
example_domain = "example123.com"

# Use main.py CLI
result = subprocess.run(
    ['python', 'main.py', 'domain', example_domain],
    cwd=str(project_root),
    capture_output=True,
    text=True
)

print("CLI Output:")
print(result.stdout)
if result.stderr:
    print("Errors:")
    print(result.stderr)
```

## Programmatic Inference Examples

Alternatively, use models directly:

```python
import pickle
from utils.feature_utils import extract_domain_features

# Load domain model
domain_model_path = project_root / 'models' / 'domain_rf_model.pkl'

if domain_model_path.exists():
    with open(domain_model_path, 'rb') as f:
        domain_model = pickle.load(f)
    
    # Predict a domain
    test_domain = "suspicious123.com"
    df_test = pd.DataFrame({'domain': [test_domain], 'label': [0]})
    features = extract_domain_features(df_test)
    X_test = features.drop('label', axis=1)
    
    prediction = domain_model.predict(X_test)[0]
    probability = domain_model.predict_proba(X_test)[0]
    
    result = "MALICIOUS" if prediction == 1 else "BENIGN"
    print(f"Domain: {test_domain}")
    print(f"Prediction: {result}")
    print(f"Probability: Benign={probability[0]:.3f}, Malicious={probability[1]:.3f}")
else:
    print(f"⚠️ Model not found: {domain_model_path}")
    print("Please train models using 02_classical_models.ipynb first")
```
